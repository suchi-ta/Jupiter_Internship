# Model Testing

```python
# Loading the model
import joblib

model_1 = joblib.load("logistic_fraud_model.pkl")
model_2= joblib.load("tuned_logistic_fraud_model.pkl")
```

```python
# load the test data
import pandas as pd

X_test_data = pd.read_csv("X_test.csv")
y_test_data = pd.read_csv("y_test.csv")
```

```python
y_pred_m1 = model_1.predict(X_test_data)
y_pred_proba_m1 = model_1.predict_proba(X_test_data)[:, 1]
y_pred_m2 = model_2.predict(X_test_data)
y_pred_proba_m2 = model_2.predict_proba(X_test_data)[:, 1]
```

# Model Evaluation

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

print("Accuracy:", accuracy_score(y_test_data, y_pred_m1))
print("Precision:", precision_score(y_test_data, y_pred_m1))
print("Recall:", recall_score(y_test_data, y_pred_m1))
print("F1 Score:", f1_score(y_test_data, y_pred_m1))
print("ROC-AUC:", roc_auc_score(y_test_data, y_pred_proba_m1))
```

```python
# Tuned model evaluation metric
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

print("Accuracy:", accuracy_score(y_test_data, y_pred_m2))
print("Precision:", precision_score(y_test_data, y_pred_m2))
print("Recall:", recall_score(y_test_data, y_pred_m2))
print("F1 Score:", f1_score(y_test_data, y_pred_m2))
print("ROC-AUC:", roc_auc_score(y_test_data, y_pred_proba_m2))
```

Accuracy is pretty high but in the case of fraud detection it can be misleading so looking at precision which is again very high which tells us that how many fruad detected were actually fraud.
Recall tells us out of all actual frauds how many we were able to detect which is also pretty high
ROC-AUC measures the models ability to distinguish between fraud and non fraud so our model is workng well

# Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

cm = confusion_matrix(y_test_data, y_pred_m1)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model_1.classes_)
disp.plot(cmap="Reds")
plt.show()
```

```python
# Tuned model results
cm = confusion_matrix(y_test_data, y_pred_m2)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model_2.classes_)
disp.plot(cmap="Blues")
plt.show()
```

- THe True Negative are non fruad transcations which were correctly classified by our model
- False positive these are wrongly classified as fraud but it is very small compared to the true positive shows that it happens rarely
- False Negative these are the missed fraud cases and these are very important as we were unable to detect these fraud
- True Positive are the correctly identified fruad cases our model is able to catch almost all frauds

We Can See that we get better results with the tuned model
